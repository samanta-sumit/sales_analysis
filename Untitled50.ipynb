```python
#updated datafram
path="C:\\Users\\91766\\Pandas-Data-Science-Tasks-master\\SalesAnalysis\\all_data.csv"
all_data= pd.read_csv(path)

all_data.head(10)
```

##clean the data

drop rows of NAN

```python
nan_df=all_data[all_data.isna().any(axis=1)]
nan_df.head()
```

```python
#drop nan value
all_data=all_data.dropna(how='all')
all_data.head()
```

```python
#find Or vale and exclude it 
all_data = all_data[all_data['Order Date'].str[0:2]!='Or']
```

```python
#covert columns inn correct type

all_data['Quantity Ordered'] = all_data['Quantity Ordered'].astype("int64") #make int
all_data['Price Each'] = pd.to_numeric(all_data['Price Each']) #make int
```

#task 4 :add a city column

```python
#lets use .apply()
def get_state(address):
          return address.split(',')[-1].split(' ')[1]

all_data['city']=all_data['Purchase Address'].apply(lambda x: x.split(',')[1]+ '' + get_state(x))


all_data.head()
```

```python
#additional columns
```

--question1:what was the best month for sales?how much was earned that month?--

```python
#months column
all_data['Month']= all_data['Order Date'].str[0:2]
all_data['Month'] = all_data['Month'].astype("int")
all_data.head()
```

```python
#add a sales column
all_data['Sales'] = all_data['Quantity Ordered']*all_data['Price Each']
all_data.head(1000)
```

```python
results=all_data.groupby('Month').sum()
```

```python
import matplotlib.pyplot as plt

months = range(1,13)

plt.bar(months,results['Sales'])
plt.xticks(months)
plt.ylabel('sales in usd($)')
plt.xlabel('Months number')
```

####question 2: whar ciry had the highest number of sales



```python

result=all_data.groupby('city').sum()
result
```

```python
city = [city for city, df in all_data.groupby('city')]

plt.bar(city,result['Sales'])
plt.xticks(city,rotation = 'vertical',size=8)
plt.ylabel('sales in usd($)')
plt.xlabel('city')
```

question 3:what time should we display ad to maximum likelyhood of customers buying product



```python
#added hours and minute cloumn
all_data['Order Date'] = pd.to_datetime(all_data['Order Date'])
```

```python
all_data['Hour']= all_data['Order Date'].dt.hour
all_data['Minute']= all_data['Order Date'].dt.minute
all_data.head()
```

```python
Hour = [hour for hour ,df in all_data.groupby('Hour')]

plt.plot(Hour, all_data.groupby(['Hour']).count())
plt.xticks(Hour)
plt.xlabel('Hour')
plt.ylabel('no of orders')
plt.grid()
```

#### question 4 :what product  sold the most?

```python
product_group=all_data.groupby('Product')
qunatity_order= product_group.sum()['Quantity Ordered']

products = [product for product, df in product_group]
plt.bar(products,qunatity_order)
plt.xticks(products, rotation='vertical',size=8)
plt.show()
```

```python

prices = all_data.groupby('Product').mean()['Price Each']

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar( products,qunatity_order, color='g')
ax2.plot(products,prices, color='b')

ax1.set_xlabel('Product Name')
ax1.set_ylabel('Quantity Ordered', color='g')
ax2.set_ylabel('Price ($)', color='b')
ax1.set_xticklabels(products, rotation='vertical', size=8)

fig.show()
```
